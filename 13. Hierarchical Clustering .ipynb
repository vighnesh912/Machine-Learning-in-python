```python
import matplotlib.pyplot as plt
import numpy as np
A = np.array([[5,3], [10,15],[15,12],[24,10],[30,30],[85,70],[71,80],[60,78],[70,55]])
```

```python
plt.scatter(A[:,0],A[:,1])
plt.show()
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
X = linkage(A, 'centroid')
dendrogram(X)
plt.show()
```

```python
X = linkage(A, 'single')
dendrogram(X)
plt.show()
```

```python
X = linkage(A, 'complete')
dendrogram(X)
plt.show()
```

```python
X = linkage(A, 'average')
dendrogram(X)
plt.show()
```
